```python
import re
from collections import Counter

#Copied from http://norvig.com/spell-correct.html
def words(text): return re.findall(r'\w+', text.lower())

WORDS = {"fog":13651,"for":7097981,"ford":15620,"fore":6699,"forge":3892,"fork":8219,"form":307506,"fort":23113,
         "frog":4119,"go":816536,"god":552668,"good":966602,"gorge":6177,"got":432016,"groan":9995,"grow":64005,
         "grown":57772,"lord":422407,"more":1899787,"nor":349691,"of":33950064,"off":545832,"or":4228287,"work":829823}

def P(word, N=sum(WORDS.values())): 
    "Probability of `word`."
    return WORDS[word] / N

def correction(word): 
    "Most probable spelling correction for word."
    return max(candidates(word), key=P)

def candidates(word): 
    "Generate possible spelling corrections for word."
    return (known([word]) or known(edits1(word)) or known(edits2(word)) or [word])

def known(words): 
    "The subset of `words` that appear in the dictionary of WORDS."
    return set(w for w in words if w in WORDS)

def edits1(word):
    "All edits that are one edit away from `word`."
    letters    = 'abcdefghijklmnopqrstuvwxyz'
    splits     = [(word[:i], word[i:])    for i in range(len(word) + 1)]
    deletes    = [L + R[1:]               for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R)>1]
    replaces   = [L + c + R[1:]           for L, R in splits if R for c in letters]
    inserts    = [L + c + R               for L, R in splits for c in letters]
    return set(deletes + transposes + replaces + inserts)

def edits2(word): 
    "All edits that are two edits away from `word`."
    return (e2 for e1 in edits1(word) for e2 in edits1(e1))
```

```python
print('A) ')
four_words = ['frog', 'gorf', 'forg','grof']
for word in four_words:
    print('correction('+str(word)+') --> ' +str(correction(word)))
```

```python
print('B) ')
six_words = ['foog', 'freg', 'frg', 'frgge', 'froot', 'froog']
for word in six_words:
    print('correction('+str(word)+') --> ' +str(correction(word)))
```

```python
from scipy.stats import linregress
from scipy.stats import pearsonr
from scipy.stats import spearmanr
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
prob1 = [0, 4, 5, 8, 5, 1, 10, 15, 5, 1, 0, 1, 5, 0, 4, 7, 7, 5, 11, 7, 15, 5, 4, 15, 7]
prob2 = [0, 10, 0, 15, 13, 6, 9, 6, 11, 15, 10, 15, 0, 10, 15, 15, 10, 11, 11, 11, 11, 4, 11, 3, 8]
prob3 = [8, 4, 8, 2, 3, 6, 5, 5, 3, 1, 5, 3, 7, 5, 1, 1, 5, 5, 5, 5, 5, 5, 2, 5, 3]
prob4 = [0, 4, 5, 8, 5, 1, 10, 0, 5, 1, 0, 2, 6, 1, 5, 6, 6, 4, 10, 6, 1, 6, 5, 1, 5]
prob5 = [0, 4, 5, 8, 5, 1, 10, 0, 5, 1, 0, 1, 5, 0, 4, 7, 7, 5, 11, 7, 0, 5, 4, 0, 7]
allprob = [prob1, prob2, prob3, prob4, prob5]
```

```python
for i in range(len(allprob)):
    j = i
    while j < len(allprob) - 1:
        pearson = pearsonr(allprob[i], allprob[j+1])
        j += 1
        print('Prob' + str(i+1) + ' and '+ 'prob'+ str(j+1) + ' Pearson r val = ' + str(pearson[0]))
        if pearson[1] < 0.05:
            print()
            print('prob' + str(i+1) + ' and' + ' prob' + str(j+1) + ' are correlated')
            print()
```

```python
for i in range(len(allprob)):
    j = i
    while j < len(allprob) - 1:
        spearman = spearmanr(allprob[i], allprob[j+1])
        j += 1
        print('Prob' + str(i+1) + ' and '+ 'prob'+ str(j+1) + ' Spearman r val = ' + str(spearman[0]))
        slope, intercept, r_value, p_value, std_err = linregress(allprob[i],allprob[j])
        if spearman[1] < 0.05:
            print()
            print('prob' + str(i+1) + ' and' + ' prob' + str(j+1) + ' are correlated')
            print()
            plt.xlabel('prob' + str(i+1), fontsize=15)
            plt.ylabel('prob' + str(j+1), fontsize=15)
            plt.plot(allprob[i], allprob[j], 'o')
            plt.plot(np.array(allprob[i]), intercept + slope*np.array(allprob[i]), 'r', label='fitted line');
            plt.show()
```

```python
def tenthousandways_uniform(trials):
    res = []
    for i in range(trials):
        sample = np.random.uniform(0,50000, size=10)
        subsets = np.zeros((1024, 10))
        for subset_i in range(1024):
            indices = list("{:010b}".format(subset_i))
            for pos in range(len(indices)):
                if indices[pos] == "1":
                    subsets[subset_i][pos] = sample[pos]

        sums = np.sum(subsets, axis=1)

        diffs = np.abs(130000 - sums)
        min_idx = np.argmin(diffs)
        res.append(diffs[min_idx])
    return res

```

```python
%%time
closest = sorted(tenthousandways_uniform(10000))
```

```python
top10 = closest[10]
top100 = closest[100]
print("99.9 percentile is " + str(top10))
print("99 percentile is " + str(top100))
if 0.28 < top10:
    print("0.28 is in the 99.9th percentile")
elif 0.28 < top100:
    print("0.28 is in the 99th percentile")
else:
    print("0.28 isn't in the 99.9 or 99th percentile")
```

```python
import scipy

def tenthousandways_gamma(trials):
    res = []
    for i in range(trials):
        #sample = np.random.uniform(0,50000, size=10)
        sample = scipy.stats.gamma.rvs(0.6, loc=1000, scale=15000, size=10)
        subsets = np.zeros((1024, 10))
        for subset_i in range(1024):
            indices = list("{:010b}".format(subset_i))
            for pos in range(len(indices)):
                if indices[pos] == "1":
                    subsets[subset_i][pos] = sample[pos]

        sums = np.sum(subsets, axis=1)

        diffs = np.abs(130000 - sums)
        min_idx = np.argmin(diffs)
        res.append(diffs[min_idx])
    return res
```

```python
closest = sorted(tenthousandways_gamma(10000))
top10 = closest[10]
top100 = closest[100]
print("99.9 percentile is " + str(top10))
print("99 percentile is " + str(top100))
if 0.28 < top10:
    print("0.28 is in the 99.9th percentile")
elif 0.28 < top100:
    print("0.28 is in the 99th percentile")
else:
    print("0.28 isn't in the 99.9 or 99th percentile")
```

```python
maleStateData = defaultdict(list)
with open('cdc1.tsv') as csvfile:
    for row in csv.reader(csvfile, delimiter='\t'):
        for state in states:
            if row[-1] == 'Unreliable': continue
            if row[0] == '' and row[1] == 'White' and row[5] == 'Not Hispanic or Latino' and row[7] == 'Male' and row[9]==state:
                maleStateData[state].append(float(row[13])/483.7)
                
femaleStateData = defaultdict(list)
with open('cdc1.tsv') as csvfile:
    for row in csv.reader(csvfile, delimiter='\t'):
        for state in states:
            if row[-1] == 'Unreliable': continue
            if row[0] == '' and row[1] == 'White' and row[5] == 'Not Hispanic or Latino' and row[7] == 'Female' and row[9]==state:
                femaleStateData[state].append(float(row[13])/218)
```

```python
maleRate = defaultdict(list)
for state in maleStateData: 
    slope, intercept, r_value, p_value, std_err = linregress(years, maleStateData[state])
    maleRate[state].append(slope)
maleRate_sorted = sorted(maleRate.items(), key=lambda x: x[1], reverse=True)
print(maleRate_sorted)
```
